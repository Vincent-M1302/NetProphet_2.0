# Zev_15

```python
from pandas import read_csv 

# read zev_15
df_zev_15_expr = read_csv('expr_zev_15_data.expr', header=None, sep='\t')
l_zev_15_expr = [v2 for v1 in df_zev_15_expr.values for v2 in v1]

# ---------------------------------- # 
# |          *** PLOT ***          | #  
# ---------------------------------- # 
import matplotlib.pyplot as plt
%matplotlib inline

# Plot Histogram
fig, axs = plt.subplots(1, 2)
n, bins, patches = axs[0].hist(l_zev_15_expr, 50, facecolor='blue')

# Plot Boxplot
box = axs[1].boxplot(l_zev_15_expr)
```
